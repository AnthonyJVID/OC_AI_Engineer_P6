```python
# Définition de l'URL et des paramètres
url = "https://world.openfoodfacts.org/cgi/search.pl"
params = {"search_terms": "champagne", "json": 1, "page_size": 21}  # "page_size": 21 pour avoir suffisamment de produits

# Requête API et vérification du résultat de la requête
response = requests.get(url, params=params)

if response.status_code == 200:
    data = response.json()
    products = data.get("products", [])

    if not products:
        print("Aucun produit trouvé.")
    else:
        # Noms des colonnes
        fieldnames = ["foodId", "label", "category", "foodContentsLabel", "image"]

        # Transformation des données JSON en CSV
        csv_data = [
            {
                "foodId": product.get("id", "N/A"),
                "label": product.get("brands", "N/A"),
                "category": product.get("categories", "N/A"),
                "foodContentsLabel": product.get("ingredients_text", "N/A"),
                "image": product.get("image_url", "N/A"),
            }
            for product in products
        ]

        # Ecriture du CSV
        with open("output.csv", "w", newline='', encoding="utf-8") as file:
            writer = csv.DictWriter(file, fieldnames=fieldnames)
            writer.writeheader()
            writer.writerows(csv_data)

        # Chargement du DataFrame et filtrage
        df = pd.read_csv("output.csv", usecols=fieldnames)
        df_filtered = df[df["foodContentsLabel"].str.contains("champagne", case=False, na=False)]

        # Filtrage des données afin d'avoir une liste complète (sans NaN)
        df_filtered.to_csv("output_final.csv", index=False)

else:
    print(f"Erreur lors de la requête API : {response.status_code}")
```

# Les grands principes des règles de protection des données personnelles.

[FR] - **R.G.P.D** => Règlement Général sur la Protection des Données    
[EN] - **G.D.P.R** => General Data Protection Regulation

- <u>1 / Le principe de finalité</u> :

Le responsable d'un fichier ne peut enregistrer et utiliser des informations sur des personnes physiques que dans un but bien précis, légal et légitime

- <u>2 / Le principe de proportionnalité et de pertinence</u> :

Les informations enregistrées doivent être pertinentes et strictement nécessaires au regard de la finalité du fichier

- <u>3 / Le principe d'une durée de conservation limitée</u> :

Il n'est pas possible de conserver des informations sur des personnes physiques dans un fichier pour une durée indéfinie.    
Une durée de conservation précise doit être fixée, en fonction du type d'information enregistrée et de la finalité du fichier

- <u>4 / Le principe de sécurité et de confidentialité</u> :

Le responsable du fichier doit garantir la sécurité des informations qu'il détient. Il doit en particulier veiller à ce que seules les personnes autorisées aient accès à ces informations

- <u>5 / Les droits des personnes</u> :

Les personnes concernées disposent de droits afin de garder la maitrise de leurs données.
